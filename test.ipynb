```python
# read XML from https://www.cbr.ru/scripts/XML_daily.asp
# <ValCurs Date="26.10.2024" name="Foreign Currency Market">
# <Valute ID="R01010">
# <NumCode>036</NumCode>
# <CharCode>AUD</CharCode>
# <Nominal>1</Nominal>
# <Name>Австралийский доллар</Name>
# <Value>64,0024</Value>
# <VunitRate>64,0024</VunitRate>
# </Valute>
# <Valute ID="R01020A">
# <NumCode>944</NumCode>
# <CharCode>AZN</CharCode>
# <Nominal>1</Nominal>
# <Name>Азербайджанский манат</Name>
# <Value>56,8622</Value>
# <VunitRate>56,8622</VunitRate>
# ...
# code:

import requests
response = requests.get('https://www.cbr.ru/scripts/XML_daily.asp')
data = response.text
from bs4 import BeautifulSoup
soup = BeautifulSoup(data, 'html.parser')
exchange_rate_dict = {}
for valute in soup.find_all('valute'):
    char_code = valute.find('charcode').text
    VunitRate = float(valute.find('vunitrate').text.replace(',', '.'))
    exchange_rate_dict[char_code] = VunitRate
```
